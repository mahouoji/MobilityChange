```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import math
import os
import csv
```

```python
is_exists = os.path.exists("../data/turnstile/date_geo")

if not is_exists:
        os.makedirs("../data/turnstile/date_geo")
        print ("Create geo dir")
```

```python
data_path = '../data/turnstile'
date_sum = os.path.join(data_path, 'date_geo')
```

```python
col_names = ['Station', 'Entries']

rideship = pd.read_csv('../data/turnstile/raw/0422.csv', names=col_names)

display(rideship)
```

```python
geo = pd.read_csv('../data/turnstile/geo/station.csv')

data = pd.merge(rideship, geo)
data['Ridership'] = data['Entries']
data.drop(columns=['Station','Entries','Borough'], inplace=True)
data.to_csv(os.path.join(date_sum, 'geo_0422.csv'), index=False, header=False)

display(data)
```

```python
data_clean = data.dropna(subset=['Latitude', 'Longitude'])
data_clean.to_csv(os.path.join(date_sum, 'geo_0422_clean.csv'), index=False, header=False)

display(data_clean)
```
